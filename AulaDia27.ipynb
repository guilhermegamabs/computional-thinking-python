```python
# Criando as matrizes A e B
A = np.array([[1, 2, 3],
              [5, 5, 9],
              [7, 34, 5]])

B = np.array([[9, 8, 7],
              [6, 5, 4],
              [3, 2, 1]])

# Calculando a soma das matrizes
soma = A + B

# Exibindo o resultado
print("Matriz A:")
print(A)
print("\nMatriz B:")
print(B)
print("\nSoma (A + B):")
print(soma)
```

```python
# Criando as matrizes A e B
A = np.array([[1, 2, 3],
              [5, 5, 9],
              [7, 34, 5]])

B = np.array([[9, 8, 7],
              [6, 5, 4],
              [3, 2, 1]])

# Calculando a multiplicação das matrizes
mul = np.multiply(A, B)
# Exibindo o resultado
print("Matriz A:")
print(A)
print("\nMatriz B:")
print(B)
print("\Multiplicação (A * B):")
print(mul)
```

```python
C = np.array([[1, 2], [3, 4], [5, 6]])

C_transposta = np.transpose(C)

# Exibindo o resultado
print("Matriz C:")
print(C)
print("\nTransposta da Matriz C:")
print(C_transposta)
```

```python
A = np.array([[1, 2, 3],
              [5, 5, 9],
              [7, 34, 5]])

A_identidade = np.identity(len(A))

# Exibindo o resultado
print("Matriz A:")
print(A)
print("\nMatriz Identidade:")
print(A_identidade)
```
