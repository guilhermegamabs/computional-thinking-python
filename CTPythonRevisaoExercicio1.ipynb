```python
# Parte 1

import pandas as pd
import numpy as np
df_pessoas = pd.DataFrame({
'nome': [' Lucas', 'mariana ', 'JOÃO', 'ana', 'Ana', 'lucas '],
'idade': [28, 32, np.nan, 23, 23, 28],
'salario': [3200, 4500, 999999, np.nan, 3000, 3200],
'cidade': ['SP', 'sp', 'SP', 'RJ', np.nan, 'sp'],
'genero': ['M', 'F', 'M', 'F', 'F', 'M'],
'escolaridade': ['Superior', 'Pós-Graduação', 'Médio', 'Superior', 'Superior', 'Superior']
})

print(df_pessoas)
```

```python
# Parte 2

# Padronizar
df_pessoas['nome'] = df_pessoas['nome'].str.strip().str.lower()
df_pessoas['cidade'] = df_pessoas['cidade'].str.strip().str.lower()

# Remover Duplicados
df_pessoas = df_pessoas.drop_duplicates()

print(df_pessoas)
```

```python
# Preencher Valores Nulos
df_pessoas['idade'] = df_pessoas['idade'].fillna(df_pessoas['idade'].mean())

print(df_pessoas)
```

```python
# Substitua o valor 999999 da coluna salario por np.nan e depois preencha os salários ausentes com a média
df_pessoas['salario'] = df_pessoas['salario'].replace(999999.0, np.nan)
df_pessoas['salario'] = df_pessoas['salario'].fillna(df_pessoas['salario'].mean())

# Calcular a média ignorando os NaNs
media_salario = df_pessoas['salario'].mean()

# Preencher os NaNs com a mediana
df_pessoas['salario'] = df_pessoas['salario'].fillna(media_salario)

print(df_pessoas)
```

```python
# 7. Aplique Label Encoding na coluna genero.
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

df_pessoas['genero'] = encoder.fit_transform(df_pessoas['genero'])

print(df_pessoas)
```

```python
# 8. Aplique One-Hot Encoding na coluna cidade.
df_pessoas = pd.get_dummies(df_pessoas, columns=['cidade'], drop_first=True)
print(df_pessoas)
```

```python
# 9. Crie uma codificação ordinal para a coluna escolaridade com a seguinte hierarquia:

from sklearn.preprocessing import OrdinalEncoder

# Ordem desejada para a escolaridade
escolaridade = ['Fundamental', 'Médio', 'Superior', 'Pós-Graduação']

# Inicializando o OrdinalEncoder com as categorias ordenadas
encoder = OrdinalEncoder(categories=[escolaridade])

# Aplicando a codificação ordinal na coluna escolaridade
df_pessoas['escolaridade'] = encoder.fit_transform(df_pessoas[['escolaridade']])

# Garantir que a coluna escolaridade seja do tipo inteiro
df_pessoas['escolaridade'] = df_pessoas['escolaridade'].astype(int)

print(df_pessoas)
```

```python
# 14. Crie uma nova coluna chamada 'salario_normalizado', onde os valores do salário estejam padronizados com média 0 e desvio padrão 1

from sklearn.preprocessing import MinMaxScaler, StandardScaler

scaler_norm = MinMaxScaler()
df_pessoas['salario_normalizado'] = scaler_norm.fit_transform(df_pessoas[['salario']])

scaler_std = StandardScaler()
df_pessoas['salario_padronizado'] = scaler_std.fit_transform(df_pessoas[['salario']])

print(df_pessoas[['salario', 'salario_normalizado', 'salario_padronizado']])
```

```python
def classificar_idade(idade):
  if idade < 30:
    return 'jovem'
  else :
    return 'adulto'

df_pessoas['faixa_etaria'] = df_pessoas['idade'].apply(classificar_idade)

print(df_pessoas[['idade', 'faixa_etaria']])
```
